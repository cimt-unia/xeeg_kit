```python
# Check Data Quality

import mne
from pathlib import Path

# RuntimeWarning
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
%matplotlib widget



# Specify the path to your concatenated FIF file
raw_file = "raw.fif"
clean_file = "cleaned.fif"

# Load raw data
raw = mne.io.read_raw_fif(raw_file, preload=True)
raw = raw.filter(l_freq=1.0, h_freq=100.0, picks='eeg', verbose=False)


# Load clean data
clean = mne.io.read_raw_fif(clean_file, preload=True)
```

```python
raw.plot(n_channels=45);
```

```python
clean.plot(n_channels=45);
```
